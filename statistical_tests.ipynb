```python
# loading the libraries
import pandas as pd
from scipy.stats import chi2_contingency, ttest_ind, mannwhitneyu
```

```python
# Loading data
df = pd.read_csv(r"data\no_show_cleaned.csv")
```

## TEST 1: Waiting Days vs No-Show

#### Hypothesis
* H₀ (null): Waiting time has no effect on no-show

* H₁ (alt): Waiting time affects no-show

```python
show = df[df['no_show'] == 0]['waiting_days']
no_show = df[df['no_show'] == 1]['waiting_days']

stat, p_value = mannwhitneyu(show, no_show)

print("P-value:", p_value)
```

#### Interpretation
There is a statistically significant relationship between waiting days and no-show (p < 0.05).

#### Conclusion
Reject H₀. Waiting time significantly affects whether a patient will no-show.

## TEST 2: Age vs No-Show
#### Hypothesis
* H₀: Age has no effect
* H₁: Age affects attendance

```python
show_age = df[df['no_show'] == 0]['age']
no_show_age = df[df['no_show'] == 1]['age']

stat, p_value = mannwhitneyu(show_age, no_show_age)

print("P-value:", p_value)
```

#### Interpretation
There is a statistically significant relationship between age and no-show (p < 0.05).

#### Conclusion
Reject H₀. Age significantly impacts appointment attendance.

## TEST 3: SMS vs No-Show 
###### (Categorical variable - Chi-Square)
#### Hypothesis
* H₀: SMS has no relationship with no-show
* H₁: SMS is related to no-show

```python
contingency = pd.crosstab(df['sms_received'], df['no_show'])

chi2, p_value, _, _ = chi2_contingency(contingency)

print("P-value:", p_value)
```

#### Interpretation
There is a statistically significant relationship between SMS reminders and no-show (p < 0.05).

#### Conclusion
Reject H₀. SMS reminders are significantly related to no-show behavior.

## TEST 4: Gender vs No-Show
#### Hypothesis
* H₀: Gender has no relationship with no-show
* H₁: Gender is related to no-show

```python
contingency = pd.crosstab(df['gender'], df['no_show'])

chi2, p_value, _, _ = chi2_contingency(contingency)

print("P-value:", p_value)
```

#### Interpretation
There is no statistically significant relationship between gender and no-show (p > 0.05).

#### Conclusion
Fail to reject H₀. Gender does not predict whether a patient will miss an appointment.

## TEST 6: Medical Conditions
#### Hypothesis
* H₀: Medical conditions (hypertension, diabetes, alcoholism) have no relationship with no-show
* H₁: Medical conditions are related to no-show

```python
for col in ['hypertension', 'diabetes', 'alcoholism']:
    contingency = pd.crosstab(df[col], df['no_show'])
    chi2, p_value, _, _ = chi2_contingency(contingency)
    print(col, "p-value:", p_value)
```

#### Interpretation
There is a statistically significant relationship between medical conditions (hypertension, diabetes, alcoholism) and no-show (p < 0.05).

#### Conclusion
Reject H₀. Medical conditions are significantly associated with no-show behavior.
